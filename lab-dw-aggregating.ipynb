1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.

```python

import pandas as pd
import numpy as np

url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv'
df = pd.read_csv(url)
df = df.dropna(how='all')
df.columns = df.columns.str.replace('ST', 'State')
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.describe()
```

```python
df["response"].value_counts()
```

```python
df_claim_check = (df["total_claim_amount"] > 1000)
df_claim_check.value_counts()
```

```python
# Yes to no ratio is about 14.2% of the responses are yes and there are 491 total claims over 1000 the outcome seems to be right. 491*0.142 = 69,722 and the frame has 67 rows.

df_yes = df[(df["total_claim_amount"] > 1000) & (df["response"] == "Yes")]
df_yes
```

2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

```python
# There seems to be no significant difference between males and females in all cattegories. The claims seems to follow a logical order. Personal auto's have the most claims,
# after we have the corporate cars and special cars to follow. 

claim_avg_gender_yes = df[df["response"] == "Yes"]
claim_avg_gender_yes.groupby(["policy_type", "gender"])["total_claim_amount"].count()
```
